```r
loads <- do.call(rbind, loads)
loads
```

```r
library(UnivRNG)
library(mvtnorm)

# generate standard deviations for noise
noise <- rgamma(n=90, shape=1, scale=0.01)

# generate noise from t distribution with df=5
noise <- rmvt(n=120, sigma=diag(noise**2), df=5)
noise*100
```

```r
# compute returns
returns <- as.data.frame(as.matrix(fact[,c('mrp', 'smb', 'hml')]) %*% t(loads) + noise)
colnames(returns) <- sprintf('Asset %d', 1:90)

returns
```

```r
# save the simulated returns in a csv file
write.csv(returns, 'returns.csv', row.names=FALSE)
```
