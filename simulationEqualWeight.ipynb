```r
df <- read.csv('returns.csv')
```

```r
# columns 1 to 15: small, low b/m
# columns 16 to 30: small, mid b/m
# columns 31 to 45: small, high b/m
# columns 46 to 60: big, low b/m
# columns 61 to 75: big, mid b/m
# columns 76 to 90: big, high b/m

head(df,15)
```

```r
# compute values of each asset (starting from 1 dollar)
values <- exp(cumsum(log(1 + df[,3:92])))
list(sl=values[,1:15], sm=values[,16:30], sh=values[,31:45], bl=values[,46:60], bm=values[,61:75], bh=values[,76:90])
```

```r
# compute return of equal weight portfolio
ewReturn <- as.data.frame(do.call(cbind, list(period=1:120, percentReturn=100*rowSums(as.matrix(df[,3:92]))/90)))
head(ewReturn,45)
```

```r
# compute value of equal weight portfolio (starting with 1 dollar)
ewValue <- as.data.frame(do.call(cbind, list(period=1:120, value=exp(cumsum(log(1+ewReturn$percentReturn/100))))))
head(ewValue,45)
```

```r
library(ggplot2)

# plot value of equal weighted portfolio
ggplot(ewValue) + geom_line(mapping=aes(x=period, y=value), colour='red') + xlab('Time') + ylab('Value of EW Portfolio') + labs(title='Value of Equal-Weighted Portfolio over Time')
```
