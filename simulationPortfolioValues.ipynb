```r
dfReturns <- read.csv('returns.csv')
df <- read.csv('portfolioReturns.csv')
```

```r
# Description of columns for dfReturns
# columns 1 to 15: small, low b/m
# columns 16 to 30: small, mid b/m
# columns 31 to 45: small, high b/m
# columns 46 to 60: big, low b/m
# columns 61 to 75: big, mid b/m
# columns 76 to 90: big, high b/m

head(df,15)
head(dfReturns,15)
```

```r
# compute values of each asset (starting from 1 dollar)
values <- exp(cumsum(log(1 + dfReturns)))
list(sl=values[,1:15], sm=values[,16:30], sh=values[,31:45], bl=values[,46:60], bm=values[,61:75], bh=values[,76:90])
```

```r
# compute return and value of equal weight portfolio
df$ewReturn <- rowSums(as.matrix(dfReturns))/90
df$ewValue <- exp(cumsum(log(1+df$ewReturn)))

# compute value of dr porfolio
df$drValue <- exp(cumsum(log(1+df$drReturn))) * df$ewValue[37]

df
```

```r
library(ggplot2)

# value of portfolios
pewValue <- (ggplot() + 
             geom_line(data=df, mapping=aes(x=period, y=ewValue, colour='.')) + 
             geom_line(data=df[37:120,], mapping=aes(x=period, y=drValue, colour='..')) +
             scale_color_discrete(name='Portfolio', labels=c('EW', 'DR')) +
             xlab('Time') + 
             ylab('Value of Portfolio') + 
             labs(title='Value of Portfolios over Time'))

# return of equal weighted portfolio
pewReturn <- (ggplot(df) + 
              geom_point(mapping=aes(x=period, y=ewReturn, colour=regimeName, shape=regimeName)) + 
              xlab('Time') + 
              ylab('Return of EW Portfolio') + 
              labs(title='Return of Equal-Weighted Portfolio over Time'))

# return of dr portfolio
pdrReturn <- (ggplot(df[37:120,]) + 
              geom_point(mapping=aes(x=period, y=drReturn, colour=regimeName, shape=regimeName)) + 
              xlab('Time') + 
              ylab('Return of DR Portfolio') + 
              labs(title='Return of DR Portfolio over Time'))
```

```r
# plot value of equal weight portfolio
pewValue
```

```r
# plot return of equal weighted portfolio
pewReturn
```

```r
# plot return of dr portfolio
pdrReturn
```

```r
# sharpe ratio of ew portfolio
ewSR <- mean(df$ewReturn)/sd(df$ewReturn)
ewSR
```

```r
# sharpe ratio of dr portfolio
drSR <- mean(df[37:120,]$drReturn)/sd(df[37:120,]$drReturn)
drSR
```
